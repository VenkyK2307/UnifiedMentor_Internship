## Building A model

### Importing Required Libraries

```python
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
import ipywidgets as widgets
from IPython.display import display
```

### Loading CSV

```python
# 📥 Load your CSV
df = pd.read_csv("Iris.csv")

# 🔧 Preprocess
X = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
y = df['Species'].astype('category').cat.codes  # convert to numbers
target_names = df['Species'].astype('category').cat.categories  # map back later
```

### Building Model and Training The Model

```python
# 🧠 Train model
model = RandomForestClassifier()
model.fit(X, y)
```

### Creating the DropDowns 

```python
# 🎛️ Create dropdowns/sliders for inputs
sepal_length = widgets.FloatSlider(value=5.0, min=4.0, max=8.0, step=0.1, description='Sepal Length')
sepal_width = widgets.FloatSlider(value=3.0, min=2.0, max=4.5, step=0.1, description='Sepal Width')
petal_length = widgets.FloatSlider(value=1.5, min=1.0, max=7.0, step=0.1, description='Petal Length')
petal_width = widgets.FloatSlider(value=0.2, min=0.1, max=2.5, step=0.1, description='Petal Width')

button = widgets.Button(description='Predict Flower 🌸')
output = widgets.Output()
```

### Prediction Function

```python
# 🔍 Prediction function
def on_button_clicked(b):
    features = [[sepal_length.value, sepal_width.value, petal_length.value, petal_width.value]]
    prediction = model.predict(features)[0]
    flower = target_names[prediction]

    with output:
        output.clear_output()
        print(f"🌼 **Predicted Flower Type:** {flower.capitalize()}")
```

### Predicton Output

```python

button.on_click(on_button_clicked)
display(sepal_length, sepal_width, petal_length, petal_width, button, output)
```
